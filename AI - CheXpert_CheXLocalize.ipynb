```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import numpy as np
import pandas as pd
from PIL import Image
import json

with open('/content/drive/MyDrive/AI_in_BME/CheXlocalize/gt_annotations_val.json') as f:
   dataset = json.load(f)
print(dataset)

# dataset = pd.DataFrame(dataset)
# dataset.iloc[10].isnull().any

# Load the CXR image
img_path = '/content/drive/MyDrive/AI_in_BME/CheXpert/val/patient64648/study1/view1_frontal.jpg'
img = Image.open(img_path)

patient_id = 'patient64648_study1_view1_frontal'
disease_name = 'Lung Lesion'
datax = dataset[patient_id][disease_name]
print(datax)

# Plot the CXR image
fig, ax = plt.subplots()
ax.imshow(np.array(img))

# Loop through the areas of interest for the CXR
for coords in datax:
    # Create a polygon patch for the area of interest
    polygon = patches.Polygon(coords, linewidth=1, edgecolor='r', facecolor='r')
    # Add the polygon patch to the plot
    ax.add_patch(polygon)
    # Add a label to the polygon
    label = 'Lung Lesion'
    ax.text(coords[0][1], coords[0][1], label, color='r')

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import numpy as np
import pycocotools.mask as mask_utils

# Load the image
image = plt.imread('/content/drive/MyDrive/AI_in_BME/CheXpert/val/patient64541/study1/view1_frontal.jpg')

# Decode the mask
counts = datax['patient64622_study1_view1_frontal']['Enlarged Cardiomediastinum']['counts']
mask = mask_utils.decode({'size': [2320, 2828], 'counts': counts})

# Create a figure and plot the image
fig, ax = plt.subplots()
ax.imshow(image)

# Add the mask as a patch
mask = np.ma.masked_where(mask == 0, mask)
ax.imshow(mask, alpha=0.5, cmap='Greens')

# Show the plot
plt.show()
```
